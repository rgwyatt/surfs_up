```python
# Dependencies
import pandas as pd
import numpy as np
import re

# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func
```

```python
engine = create_engine("sqlite:///hawaii.sqlite")

# reflect an existing database into a new model
Base = automap_base()
# reflect the tables
Base.prepare(engine, reflect=True)

# Save references to each table
Measurement = Base.classes.measurement
Station = Base.classes.station
```

```python
# Create our session (link) from Python to the DB
session = Session(engine)
```

## D1: Determine the Summary Statistics for June

```python
# 1. Import the sqlalchemy extract function.
from sqlalchemy import extract

# 2. Write a query that filters the Measurement table to retrieve the temperatures for the month of June. 
june10_list = session.query(Measurement.tobs).\
    filter(Measurement.date >= '2010-06-01').\
    filter(Measurement.date <= '2010-06-30').all()
june11_list = session.query(Measurement.tobs).\
    filter(Measurement.date >= '2011-06-01').\
    filter(Measurement.date <= '2011-06-30').all()
june12_list = session.query(Measurement.tobs).\
    filter(Measurement.date >= '2012-06-01').\
    filter(Measurement.date <= '2012-06-30').all()
june13_list = session.query(Measurement.tobs).\
    filter(Measurement.date >= '2013-06-01').\
    filter(Measurement.date <= '2013-06-30').all()
june14_list = session.query(Measurement.tobs).\
    filter(Measurement.date >= '2014-06-01').\
    filter(Measurement.date <= '2014-06-30').all()
june15_list = session.query(Measurement.tobs).\
    filter(Measurement.date >= '2015-06-01').\
    filter(Measurement.date <= '2015-06-30').all()
june16_list = session.query(Measurement.tobs).\
    filter(Measurement.date >= '2016-06-01').\
    filter(Measurement.date <= '2016-06-30').all()
june17_list = session.query(Measurement.tobs).\
    filter(Measurement.date >= '2017-06-01').\
    filter(Measurement.date <= '2017-06-30').all()
```

```python
#  3. Convert the June temperatures to a list.
june_temps = []
june_temps.append(june10_list)
june_temps.append(june11_list)
june_temps.append(june12_list)
june_temps.append(june13_list)
june_temps.append(june14_list)
june_temps.append(june15_list)
june_temps.append(june16_list)
june_temps.append(june17_list)
print(june_temps)
```

```python
# 4. Create a DataFrame from the list of temperatures for the month of June. 
june_df = pd.DataFrame(june_temps)
june_df.head()
```

```python
# 5. Calculate and print out the summary statistics for the June temperature DataFrame.
june_df.describe()
```
